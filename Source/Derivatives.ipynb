```python
import matplotlib.pyplot as plt
import numpy as np
from findiff import FinDiff
from numpy import sin
from numpy import pi as PI
```

```python
def f(x):
    return sin(x)/x
```

```python
x = np.linspace(-10, 10, 100)
y = f(x)

# Define the derivative:
d_dx = FinDiff(0, x, 1)

# Apply it:
df_dx = d_dx(y) 
```

```python
plt.plot(x,y)
plt.plot(x,df_dx)
plt.xlim(-10,10)
plt.ylim(-1,1)
plt.grid(True)
plt.show()
```

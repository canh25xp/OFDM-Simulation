```python
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
from numpy import sin
from numpy import pi as PI
```

```python
save_fig = True # To save plot during simulation
save_folder = Path("results") # Where to save the figures
save_format = "pdf" # Save as what type
def Save_fig(save_name : str):
    plt.savefig(save_folder/(save_name+"."+save_format), bbox_inches='tight')
```

```python
def f(x):
    return sin(x)/x
```

```python
x = np.linspace(-10*PI, 10*PI, 1000)
y = f(x)
y1 = f(x-1*PI)
y2 = f(x-2*PI)
y3 = f(x-3*PI)
```

```python
plt.plot(x,y)
plt.xlim(-6*PI,6*PI);plt.ylim(-0.25,1.2)
plt.xlabel('Frequency');plt.ylabel('Power spectral density')
plt.grid(True)
Save_fig("single_carrier")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlim(-6*PI,6*PI);plt.ylim(-0.25,1.2)
plt.xlabel('Frequency');plt.ylabel('Power spectral density')
plt.grid(True)
Save_fig("four_subcarriers")
plt.show()
```
